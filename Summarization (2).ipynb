# Using Inference

```python
import os
from huggingface_hub import InferenceClient

# Load the Hugging Face token from Colab Secrets
from google.colab import userdata
os.environ["HF_TOKEN"] = userdata.get("HF_TOKEN")

client = InferenceClient(
    provider="hf-inference",
    api_key=os.environ["HF_TOKEN"],
)
article="""
A lonely willow tree stood by the river’s edge.
Its branches hung low, bowing to the flowing water.
The forest was busy, but no one visited the willow.
Birds preferred the sturdy oak for building their nests.
Small creatures favored the berry bushes nearby.
The willow felt forgotten in the golden spring sunshine.
One windy morning, a small blue butterfly appeared.
It was struggling, wings battered from the strong wind.
The butterfly landed gently on a low willow branch.
The willow whispered, "Rest here, little friend."
The tree tucked the butterfly securely in its leaves.
The wind blew harder, but the tree kept it safe.
The butterfly stayed for hours, enjoying the shelter.
When the storm passed, it didn't fly away immediately.
"Thank you," the butterfly said, fluttering its wings.
"I have no home, and I am far from my family."
The willow replied, "You are welcome to stay here."
So, the butterfly made a home in the willow tree.
Soon, other insects came to visit the kind willow.
Birds decided to rest on its branches during the day.
The willow was no longer lonely, but very happy.
It whispered stories to the flowing river below.
The forest animals often came to rest in its shade.
The butterfly sang stories of its travel to the tree.
Through friendship, the silent tree found its voice.
It became the most vibrant spot in the forest.
The seasons changed, but the friendship remained strong.
The willow learned that one friend makes a difference.
It taught everyone that kindness brings happiness back.
The lonely willow was lonely no more.
"""
result = client.summarization(
    article,    model="facebook/bart-large-cnn",
)
```

```python
result.summary_text
```

# Using Transformers

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn")


inputs = tokenizer([article], max_length=1024, return_tensors='pt')

summary_ids = model.generate(
    inputs['input_ids'],
    num_beams=4,
    min_length=30,
    max_length=100,
    early_stopping=True
)

result = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
```
